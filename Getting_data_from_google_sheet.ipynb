```python
from google.colab import auth
auth.authenticate_user()

import gspread
from google.auth import default
creds, _ = default()
```

```python
gc = gspread.authorize(creds)

worksheet = gc.open('trade_history_3445086').sheet1

rows = worksheet.get_all_values()
print(rows)
```

```python
# Convert to a DataFrame and render.
import pandas as pd
df_sheet = pd.DataFrame.from_records(rows)
```

```python
df_sheet.head()
```

```python
df_sheet.columns = ['DATE','STOCK','TRADE','PRICE','QTY','REASON','RETURNS']
df_sheet = df_sheet.drop([0])
df_sheet.DATE = pd.to_datetime(df_sheet['DATE'])
```
